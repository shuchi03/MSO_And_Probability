```python
from tkinter import *
from tkinter import ttk  

def superposeStutterLess(string1, string2, voc1, voc2):
    if len(string1) == len(string2) == 0:
        return [string1]
    else:
        temp = []
        if len(string1) > 0:
            h1 = string1[0]
            t1 = string1[1:]
            if len(string2) > 0:
                h2 = string2[0]
                t2 = string2[1:]
                h3 = h1.union(h2)
                h1v2 = h1.intersection(voc2)
                if h1v2.issubset(h2):
                    h2v1 = h2.intersection(voc1)
                    if h2v1.issubset(h1):
                        for res in superposeStutterLess(t1,t2,voc1,voc2):
                            temp.append([h3] + res)
                        for res in superposeStutterLess(string1,t2,voc1,voc2):
                            temp.append([h3] + res)
                        for res in superposeStutterLess(t1,string2,voc1,voc2):
                            temp.append([h3] + res)     
        return temp

def superStutterLess(s1,s2):
    return superposeStutterLess(s1,s2,voc(s1),voc(s2))

def red(str,set): 
    if len(str) == 0:
        return str
    else:
        h1 = str[0].intersection(set)
        return [h1] + red(str[1:],set)

def bc(str): 
    if len(str) < 2:
        return str
    else:
        h,t = str[0],str[1:]
        if h == t[0]:
            return bc(t)
        else:
            return [h] + bc(t)

def allenStutterLess(x,y,r):
    e = set({})
    if r == "b":
        return [e, {x}, e, {y}, e]
    elif r == "o":
        return [e, {x}, {x,y}, {y}, e]
    elif r == "m":
        return [e, {x}, {y}, e]
    elif r == "d":
        return [e, {y}, {x,y}, {y}, e]
    elif r == "s":
        return [e, {x,y}, {y}, e]
    elif r == "f":
        return [e, {y}, {x,y}, e]
    if r == "bi":
        return [e, {y}, e, {x}, e]
    elif r == "oi":
        return [e, {y}, {x,y}, {x}, e]
    elif r == "mi":
        return [e, {y}, {x}, e]
    elif r == "di":
        return [e, {x}, {x,y}, {x}, e]
    elif r == "si":
        return [e, {x,y}, {x}, e]
    elif r == "fi":
        return [e, {x}, {x,y}, e]
    elif r == "eq":
        return [e, {x,y}, e]    
    
def allInvStutterLess(string):
    e = set({})
    if string == [e, {'i'}, e, {'k'}, e]:
        return "b"
    elif string == [e, {'i'}, {'i','k'}, {'k'}, e]:
        return "o"
    elif string == [e, {'i'}, {'k'}, e]:
        return "m"
    elif string == [e, {'k'}, {'i','k'}, {'k'}, e]:
        return "d"
    elif string == [e, {'i','k'}, {'k'}, e]:
        return "s"
    elif string == [e, {'k'}, {'i','k'}, e]:
        return "f"
    elif string == [e, {'i'}, e, {'k'}, e]:
        return "bi"
    elif string == [e, {'k'}, {'i','k'}, {'i'}, e]:
        return "oi"
    elif string == [e, {'k'}, {'i'}, e]:
        return "mi"
    elif string == [e, {'i'}, {'i','k'}, {'i'}, e]:
        return "di"
    elif string == [e, {'i','k'}, {'i'}, e]:
        return "si"
    elif string == [e, {'i'}, {'i','k'}, e]:
        return "fi"
    elif string == [e, {'i','k'}, e]:
        return "eq"
    
def ttStutterLess(r1,r2):
    temp = []
    for s in superStutterLess(allenStutterLess('i','j',r1), allenStutterLess('j','k',r2)):
        temp.append(allInvStutterLess(bc(red(s,{'i','k'}))))
    return temp


def superpose(string1, string2, voc1, voc2):
    if len(string1) == len(string2) == 0:
        return [string1]
    else:
        temp = []
        if len(string1) > 0:
            h1 = string1[0]
            t1 = string1[1:]
            h1v2 = h1.intersection(voc2)
            if len(h1v2) == 0:
                for res in superpose(t1,string2,voc1,voc2):
                    temp.append([h1] + res)     
            if len(string2) > 0:
                h2 = string2[0]
                t2 = string2[1:]
                h3 = h1.union(h2)
                if h1v2.issubset(h2):
                    h2v1 = h2.intersection(voc1)
                    if h2v1.issubset(h1):
                        for res in superpose(t1,t2,voc1,voc2):
                            temp.append([h3] + res)
        if len(string2) > 0:
            h2 = string2[0]
            t2 = string2[1:]
            h2v1 = h2.intersection(voc1)
            if len(h2v1) == 0:
                for res in superpose(string1,t2,voc1,voc2):
                    temp.append([h2] + res)
        return temp

def super(s1,s2):
    return superpose(s1,s2,voc(s1),voc(s2))

def voc(string): 
    if len(string) == 0:
        return {}
    else:
        return string[0].union(voc(string[1:]))

def proj(string,set): 
    if len(string) == 0:
        return string
    else:
        h1 = string[0].intersection(set)
        if len(h1) == 0:
            return proj(string[1:],set)
        else:
            return [h1] + proj(string[1:],set)

def l(x):
    return "l" + str(x)

def r(x):
    return "r" + str(x)

def allen(x,y,rel):
    lx = l(x)
    rx = r(x)
    ly = l(y)
    ry = r(y)
    if rel == "b":
        return [{lx}, {rx}, {ly}, {ry}]
    elif rel == "o":
        return [{lx}, {ly}, {rx}, {ry}]
    elif rel == "m":
        return [{lx}, {rx,ly}, {ry}]
    elif rel == "d":
        return [{ly}, {lx}, {rx}, {ry}]
    elif rel == "s":
        return [{lx,ly}, {rx}, {ry}]
    elif rel == "f":
        return [{ly}, {lx}, {rx,ry}]
    if rel == "bi":
        return [{ly}, {ry}, {lx}, {rx}]
    elif rel == "oi":
        return [{ly}, {lx}, {ry}, {rx}]
    elif rel == "mi":
        return [{ly}, {ry,lx},{rx}]
    elif rel == "di":
        return [{lx}, {ly}, {ry}, {rx}]
    elif rel == "si":
        return [{lx,ly}, {ry}, {rx}]
    elif rel == "fi":
        return [{lx}, {ly}, {rx,ry}]
    elif rel == "eq":
        return [{lx,ly},{rx,ry}]    
    
def allInv(string):
    lx = l('i')
    rx = r('i')
    ly = l('k')
    ry = r('k')
    if string == [{lx}, {rx}, {ly}, {ry}]:
        return "b"
    elif string == [{lx}, {ly}, {rx}, {ry}]:
        return "o"
    elif string == [{lx}, {rx,ly}, {ry}]:
        return "m"
    elif string == [{ly}, {lx}, {rx}, {ry}]:
        return "d"
    elif string == [{lx,ly}, {rx}, {ry}]:
        return "s"
    elif string == [{ly}, {lx}, {rx,ry}]:
        return "f"
    elif string == [{ly}, {ry}, {lx}, {rx}]:
        return "bi"
    elif string == [{ly}, {lx}, {ry}, {rx}]:
        return "oi"
    elif string == [{ly}, {ry,lx},{rx}]:
        return "mi"
    elif string == [{lx}, {ly}, {ry}, {rx}]:
        return "di"
    elif string ==  [{lx,ly}, {ry}, {rx}]:
        return "si"
    elif string == [{lx}, {ly}, {rx,ry}]:
        return "fi"
    elif string == [{lx,ly},{rx,ry}]:
        return "eq"
    
def tt(r1,r2):
    temp = []
    for s in super(allen('i','j',r1), allen('j','k',r2)):
        temp.append(allInv(proj(s,{l('i'),r('i'),l('k'),r('k')})))
    return temp

def showWithBorder(r1,r2,root):
    s1 = allen('i','j',r1)
    s2 = allen('j','k',r2)
    show_input_strings_with_border(root, s1, s2)
    for i,s in enumerate(super(s1,s2)):
        f = Frame(root, height=20, width=20)
        f.pack_propagate(0) # don't shrink
        f.place(x=680, y=145+((i+1)*50))
        label = Label(f, text=allInv(proj(s,{l('i'),r('i'),l('k'),r('k')}))) #label
        label.pack(fill=BOTH, expand=1)
        show_output_string_with_border(root,s,i+1)
        
def showStutterless(r1,r2,root):
    s1 = allenStutterLess('i','j',r1)
    s2 = allenStutterLess('j','k',r2)
    show_input_strings_stutterless(root, s1, s2)
    for i,s in enumerate(superStutterLess(s1,s2)):
        f = Frame(root, height=20, width=20)
        f.pack_propagate(0) # don't shrink
        f.place(x=0, y=145+((i+1)*50))
        label = Label(f, text=allInvStutterLess(bc(red(s,{'i','k'}))))#label
        label.pack(fill=BOTH, expand=1)
        show_output_string_stutterless(root,s,i+1)

def retrieve_relation(root):  
    clear_canvas(root)
    
    Label(root, text="Interior").place(x=20, y=85)  #label
    ttk.Separator(root).place(x=0, y=175, relwidth=1)
    Label(root, text="Border").place(x=700, y=85)  #label
    r1= firstRelation.get()
    r2= secondRelation.get()
    showWithBorder(r1, r2,root)
    showStutterless(r1, r2,root)
    
    
def show_input_strings_with_border(root, s1, s2):
    
    for i, s in enumerate(s1):
        display_str = str(s).replace('{', '').replace('}', '').replace('\'', '')
        f = Frame(root, height=20, width=90)
        f.pack_propagate(0) # don't shrink
        f.place(x=700+90*i, y=110)
        label = Label(f, text=display_str, borderwidth=1, relief="solid", bg='linen')  #label
        label.pack(fill=BOTH, expand=1)
    for i, s in enumerate(s2):
        display_str = str(s).replace('{', '').replace('}', '').replace('\'', '')
        f = Frame(root, height=20, width=90)
        f.pack_propagate(0) # don't shrink
        f.place(x=700+90*i, y=145)
        label = Label(f, text=display_str, borderwidth=1, relief="solid", bg='linen')  #label
        label.pack(fill=BOTH, expand=1)
        
    
def show_output_string_with_border(root,s3,rowNum):
    for i, s in enumerate(s3):
        display_str = str(s).replace('{', '').replace('}', '').replace('\'', '')
        f = Frame(root, height=20, width=90)
        f.pack_propagate(0) # don't shrink
        f.place(x=700+90*i, y=145+rowNum*50)
        label = Label(f, text=display_str, borderwidth=1, relief="solid", bg='linen')  #label
        label.pack(fill=BOTH, expand=1)
        
def show_input_strings_stutterless(root, s1, s2):
    
    for i, s in enumerate(s1):
        f = Frame(root, height=20, width=90)
        f.pack_propagate(0) # don't shrink
        f.place(x=20+90*i, y=110)
        if(s == set()):
            s=" "
        display_str = str(s).replace('{', '').replace('}', '').replace('\'', '')
        label = Label(f, text=display_str, borderwidth=2, relief="raised", bg="lavender blush")  #label
        label.pack(fill=BOTH, expand=1)
    for i, s in enumerate(s2):
        
        f = Frame(root, height=20, width=90)
        f.pack_propagate(0) # don't shrink
        f.place(x=20+90*i, y=145)
        if(s == set()):
            s=" "
        display_str = str(s).replace('{', '').replace('}', '').replace('\'', '')
        label = Label(f, text=display_str, borderwidth=2, relief="raised", bg="lavender blush")  #label
        label.pack(fill=BOTH, expand=1)
        
    
def show_output_string_stutterless(root,s3,rowNum):
    for i, s in enumerate(s3):
        f = Frame(root, height=20, width=90)
        f.pack_propagate(0) # don't shrink
        f.place(x=20+90*i, y=145+rowNum*50)
        if(s == set()):
            s=" "
        display_str = str(s).replace('{', '').replace('}', '').replace('\'', '')
        label = Label(f, text=display_str, borderwidth=2, relief="raised", bg="lavender blush")  #label
        label.pack(fill=BOTH, expand=1)
    
def clear_canvas(frame):
    for widget in frame.winfo_children():
        widget.destroy()
    

root = Tk()
root.geometry("1500x1000+15+15")
root.title("Allen Interval Relations")
canvas = Canvas(root)
firstRelation = StringVar()
secondRelation = StringVar()
Label(root, text="Relation between i and j").grid(row=0, padx=(10, 10))  #label
Entry(root, textvariable = firstRelation, bg="LightSteelBlue1", borderwidth=1, relief="solid").grid(row=0, column=3) #entry textbox
Label(root, text="Relation between j and k").grid(row=1, padx=(10, 10))  #label
Entry(root, textvariable = secondRelation, bg="LightSteelBlue1", borderwidth=1, relief="solid").grid(row=1, column=3) #entry textbox


fr = Frame(root, height=600, width=1350)
fr.pack_propagate(0) # don't shrink
fr.place(x=10, y=60)
WSignUp = Button(root, text="Get Transitivity Relation", command= lambda: retrieve_relation(fr)).grid(row=3, column=0, padx=(10, 10), sticky=W)

button = Button(root, text="Clear", command= lambda: clear_canvas(fr)).grid(row=4, column=0, padx=(10, 10),  sticky=W)
root.mainloop()
```
