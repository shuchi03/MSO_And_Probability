```python
from tkinter import *
from tkinter import font as tkFont
import math
import  tkinter as tk
'''
Superposition implementation on the given two strings
'''
def superpose(string1, string2, voc1, voc2):
    if len(string1) == len(string2) == 0:
        return [string1]
    else:
        temp = []
        if len(string1) > 0:
            h1 = string1[0]
            t1 = string1[1:]
            h1v2 = h1.intersection(voc2)
            if len(h1v2) == 0:
                for res in superpose(t1,string2,voc1,voc2):
                    temp.append([h1] + res)     
            if len(string2) > 0:
                h2 = string2[0]
                t2 = string2[1:]
                h3 = h1.union(h2)
                if h1v2.issubset(h2):
                    h2v1 = h2.intersection(voc1)
                    if h2v1.issubset(h1):
                        for res in superpose(t1,t2,voc1,voc2):
                            temp.append([h3] + res)
        if len(string2) > 0:
            h2 = string2[0]
            t2 = string2[1:]
            h2v1 = h2.intersection(voc1)
            if len(h2v1) == 0:
                for res in superpose(string1,t2,voc1,voc2):
                    temp.append([h2] + res)
        return temp

def super(s1,s2):
    return superpose(s1,s2,voc(s1),voc(s2))

def voc(string): 
    if len(string) == 0:
        return {}
    else:
        return string[0].union(voc(string[1:]))

def proj(string,set): 
    if len(string) == 0:
        return string
    else:
        h1 = string[0].intersection(set)
        if len(h1) == 0:
            return proj(string[1:],set)
        else:
            return [h1] + proj(string[1:],set)

def l(x):
    return "l" + str(x)

def r(x):
    return "r" + str(x)

def allen(x,y,rel):
    lx = l(x)
    rx = r(x)
    ly = l(y)
    ry = r(y)
    if rel == "b":
        return [{lx}, {rx}, {ly}, {ry}]
    elif rel == "o":
        return [{lx}, {ly}, {rx}, {ry}]
    elif rel == "m":
        return [{lx}, {rx,ly}, {ry}]
    elif rel == "d":
        return [{ly}, {lx}, {rx}, {ry}]
    elif rel == "s":
        return [{lx,ly}, {rx}, {ry}]
    elif rel == "f":
        return [{ly}, {lx}, {rx,ry}]
    if rel == "bi":
        return [{ly}, {ry}, {lx}, {rx}]
    elif rel == "oi":
        return [{ly}, {lx}, {ry}, {rx}]
    elif rel == "mi":
        return [{ly}, {ry,lx},{rx}]
    elif rel == "di":
        return [{lx}, {ly}, {ry}, {rx}]
    elif rel == "si":
        return [{lx,ly}, {ry}, {rx}]
    elif rel == "fi":
        return [{lx}, {ly}, {rx,ry}]
    elif rel == "eq":
        return [{lx,ly},{rx,ry}]    
    
def allInv(string,s1,s2):
    lx = l(s1)
    rx = r(s1)
    ly = l(s2)
    ry = r(s2)
    if string == [{lx}, {rx}, {ly}, {ry}]:
        return "b"
    elif string == [{lx}, {ly}, {rx}, {ry}]:
        return "o"
    elif string == [{lx}, {rx,ly}, {ry}]:
        return "m"
    elif string == [{ly}, {lx}, {rx}, {ry}]:
        return "d"
    elif string == [{lx,ly}, {rx}, {ry}]:
        return "s"
    elif string == [{ly}, {lx}, {rx,ry}]:
        return "f"
    elif string == [{ly}, {ry}, {lx}, {rx}]:
        return "bi"
    elif string == [{ly}, {lx}, {ry}, {rx}]:
        return "oi"
    elif string == [{ly}, {ry,lx},{rx}]:
        return "mi"
    elif string == [{lx}, {ly}, {ry}, {rx}]:
        return "di"
    elif string ==  [{lx,ly}, {ry}, {rx}]:
        return "si"
    elif string == [{lx}, {ly}, {rx,ry}]:
        return "fi"
    elif string == [{lx,ly},{rx,ry}]:
        return "eq"

'''
Computes the transitive relations for the given two relations
'''
def transitivity_relation(r1,r2):
    temp = []
    superpose_temp=[]
    for s in super(allen(0,1,r1), allen(1,2,r2)):
        superpose_temp.append(s)
        temp.append(allInv(proj(s,{l(0),r(0),l(2),r(2)}), 0, 2))
    return temp,superpose_temp

'''
Computes the transitivity table for all the allen relations
'''
def compute_transitivty_table():
    transitivity_array = []
    rowNumber = 0
    for row in range(14):
        columnNumber = 0
        rowNumber = rowNumber + 1
        temp = []
        super_temp=[]
        for col in range(14):
            columnNumber = columnNumber + 1
            if rowNumber == 1 and columnNumber == 1:
                temp.append("Allen \nRelations")
                super_temp.append("Allen \nRelations")
            elif rowNumber == 1 and columnNumber != 1:
                temp.append(all_relations[columnNumber-2])
                super_temp.append(all_relations[columnNumber-2])
            elif columnNumber == 1 and rowNumber != 1:
                temp.append(all_relations[rowNumber-2])
                super_temp.append(all_relations[rowNumber-2])
            else:
                temp1,supertemp1=transitivity_relation(all_relations[rowNumber-2], all_relations[columnNumber-2])
                temp.append(temp1)
                super_temp.append(supertemp1)
        superpose_result.append(super_temp)
        transitivity_array.append(temp)
    return transitivity_array 
   
def possible_s_words():
    global possible_short_representations
    all_allen_possibilities = super([{'l0'}, {'r0'}],[{'l1'}, {'r1'}])
    allen_for_representation = []
    for rep in possible_short_representations:
        temp = []
        for allen in all_allen_possibilities:
            for s in super(allen, rep):
                temp.append(allInv(proj(s,{l(0),r(0),l(1),r(1)}), 0, 1))
        allen_for_representation.append(temp)
    return allen_for_representation
        
def get_shorter_s_word(relation_array):
    global shorter_representations, possible_short_representations
    i = 0
    while(i<len(shorter_representations)):
        if(set(relation_array) == set(shorter_representations[i])):
            return possible_short_representations[i]
        i = i+1
    
'''
Drawing the transitivity grid on the canvas along with the text inside each box
'''
def draw_grid(view, selected_row, selected_column):
    w=1400
    h=600
    gridWidth = w / 14
    gridHeight = h / 14
    rowNumber = 0
    for row in range(14):
        columnNumber = 0
        rowNumber = rowNumber + 1
        for col in range(14):
            columnNumber = columnNumber + 1
            if rowNumber == 1 and columnNumber == 1:
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill ='lavender',outline='white')
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
            elif rowNumber == 1 and columnNumber != 1:
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill ='lavender',outline='white')
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
            elif columnNumber == 1 and rowNumber != 1:
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill ='lavender',outline='white')
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
            elif selected_row == rowNumber and selected_column == columnNumber:
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill ='lavender',outline='white')
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
                if len(list(rect_text)) == 13:
                    rect_text = "All \nrelations"
                elif(len(list(rect_text)) > 4):
                    rect_text = (str(list(rect_text)[0:4]) + '\n' + str(list(rect_text)[4:])).replace(',','').replace('\'','').replace('[','').replace(']','')
            else:
                rect = view.create_rectangle(col * gridWidth,row * gridHeight,(col + 1) * gridWidth,(row + 1) * gridHeight,fill ='lavender blush',outline='white')
                rect_text= transitivity_array[rowNumber-1][columnNumber-1]
                if len(list(rect_text)) == 13:
                    rect_text = "All \nrelations"
                elif(len(list(rect_text)) > 4):
                    rect_text = (str(list(rect_text)[0:4]) + '\n' + str(list(rect_text)[4:])).replace(',','').replace('\'','').replace('[','').replace(']','')
            view.itemconfig(rect, tags=(str(rowNumber), str(columnNumber)))
            t = view.create_text(col * gridWidth + 50,row * gridHeight+20, text=rect_text, font=("Courier", 13))
            view.pack()
            
'''
Clears the given frame
'''
def clear_frame(frame):
    for widget in frame.winfo_children():
        widget.destroy()            

'''
Showing the s-word in the frame 
'''        
def show_s_word(root, word):
    if sword_fr is not None:
        clear_frame(sword_fr)
    for i, s in enumerate(word):
        display_str = str(s).replace('{', '').replace('}', '').replace('\'', '')
        f = Frame(root, height=30, width=70)
        f.pack_propagate(0) 
        f.place(x=5+70*i, y=10)
        label = Label(f, text=display_str, borderwidth=2, relief="ridge", bg='lavender')  #label
        label.pack(fill=BOTH, expand=1)
    
'''
Click even on the transitivity table
'''    
def click_on_box(event):
    global previous,s_words_info_label, fr, view, sword_fr,result,choice,len_n,length_fr,prob_fr,prob_choice_fr
    choice.set(0)
    len_n.set(0)
    s_words_info_label['text'] = " "
    prob_info_label['text'] = " "
    view.pack()
    if view.find_withtag(CURRENT) and len(view.gettags(CURRENT)) == 3:
        (row, column, tag) = view.gettags(CURRENT)
        if(int(row) != 1 and int(column) != 1):
            view.itemconfig(CURRENT, fill="lavender")
            prob_info_label_text.set("Probablity information: ")
            
            prob_choice_fr = Frame(fr, height=600, width=250)
            prob_choice_fr.pack_propagate(0) 
            prob_choice_fr.place(x=0, y=45)
            
            length_fr = Frame(prob_choice_fr, height=100, width=110)
            length_fr.pack_propagate(0) 
            length_fr.place(x=0, y=100)
            
            prob_fr = Frame(fr, height=600, width=1000)
            prob_fr.pack_propagate(0) 
            prob_fr.place(x=200, y=45)
            
            Label(prob_choice_fr, text="Enter Choice to select \n n in n-world", font=("Courier", 13), justify = tk.LEFT).place(x=0,y=0)
            Radiobutton(prob_choice_fr, text="Select max length from \nsuperposed strings", font=("Courier", 13), variable=choice, value=1, command=lambda: sel(row,column)).place(x=0,y=30)
            Radiobutton(prob_choice_fr, text="Enter n", font=("Courier", 13), variable=choice, value=2, command=lambda: sel(row,column)).place(x=0,y=65)

            if previous is not None:
                draw_grid(view, int(row), int(column))
            previous = view.gettags(CURRENT)

'''
Calculating probability
'''           
def sel(row,column):
    global choice,len_n,length_fr,prob_fr,prob_choice_fr
    if int(choice.get())==2:
        Entry(length_fr, textvariable = str(len_n), width=7, bg="LightSteelBlue1", borderwidth=1, relief="solid").grid(row=0, column=0)
    else: 
        clear_canvas(length_fr)
    clear_canvas(prob_fr)
    btn_get_prob = Button(length_fr, text="Get Probabilities", command=lambda: get_prob(row,column), font=("Courier", 13)).grid(row=6, column=0, padx=(5, 5),  sticky=W)
    
def get_prob(row,column):
    global prob_fr,result
    max_length_info = StringVar()
    result = superpose_result[int(row)-1][int(column)-1]
    if int(choice.get())==2:
        max = maxLength()
        if int(len_n.get()) < max:
            clear_canvas(prob_fr)
            Label(prob_fr, textvariable=max_length_info, font=("Courier", 13)).place(x=80,y=30)
            max_length_info.set("Enter n greater than or equal to "+str(max))
        else: 
            clear_canvas(prob_fr)
            showProbabilities(all_relations[int(row)-2],all_relations[int(column)-2],prob_fr)
    else:
        result = superpose_result[int(row)-1][int(column)-1]
        showProbabilities(all_relations[int(row)-2],all_relations[int(column)-2],prob_fr)
        
def nCr(n,r):
    f = math.factorial
    return f(n) // f(r) // f(n-r)
       
def maxLength():
    max=0
    for s in result:
        if max < len(s):
            max=len(s)
    return max
    
def calculateProbR1R2(max):
    totLen = 0
    for s in result:
        totLen = totLen + nCr(max,len(s))
    return totLen

def showProbabilities(r1,r2,root):
    global choice, len_n
    if int(choice.get())==2:
        max = int(len_n.get())
    else:
        max = maxLength()
    totLen = calculateProbR1R2(max)
    for i,s in enumerate(result):
        
        f = Frame(root, height=20, width=200)
        f.pack_propagate(0) # don't shrink
        if(i>5):
            f.place(x=400, y=0+((j)*20))
        else:
            j=0
            f.place(x=0, y=0+((i)*20))
        prob = 8 / (pow(max,3)*pow(max-1,3))
        prob1 = prob * nCr(max,len(s))
        label = Label(f, text="p("+r1+","+r2+","+allInv(proj(s,{l(0),r(0),l(2),r(2)}),0,2)+") = "+str(round(prob1, 4)),font=("Courier", 13))#label
        label.pack(fill=BOTH, expand=1)
        
        f1 = Frame(root, height=20, width=200)
        f1.pack_propagate(0) # don't shrink
        if(i>5):
            f1.place(x=600, y=0+((j)*20))
            j=j+1
        else:
            f1.place(x=200, y=0+((i)*20))
        prob = nCr(max,len(s)) / totLen
        label = Label(f1, text="p("+allInv(proj(s,{l(0),r(0),l(2),r(2)}),0,2)+"|"+r1+","+r2+") = "+str(round(prob, 4)),font=("Courier", 13))#label
        label.pack(fill=BOTH, expand=1)
        
'''
Clear Frame
'''
def clear_canvas(frame):
    for widget in frame.winfo_children():
        widget.destroy()

'''
Initializing tkinter and canvas
'''
root = Tk()
root.geometry("1500x1000+15+15")
root.title("Allen Transitivity Grid")
view = Canvas(root, width=1400, height=700)
view.pack(side=TOP, fill=BOTH,padx=3,pady=3)
all_relations = ['b','bi','d','di','eq','f','fi','m','mi','o','oi','s','si']
possible_short_representations = [[{'l0'},{'l1'}],     [{'l0'},{'r1'}],  
                            [{'l1'},{'l0'}],     [{'r1'},{'l0'}],
                            [{'l0', 'l1'}],      [{'l0', 'r1'}],  
                            [{'r0'},{'l1'}],     [{'r0'},{'r1'}],  
                            [{'l1'},{'r0'}],     [{'r1'},{'r0'}],
                            [{'r0','l1'}],     [{'r0','r1'}],  
                            [{'l1'},{'r0'},{'r1'}],
                            [{'l1'},{'l0'},{'r1'}],
                            [{'l0'},{'r1'},{'r0'}],
                            [{'l0'},{'l1'},{'r0'}]]
other_representations = [['b', 'm', 'o'], ['bi','mi','oi'], ['d','di','eq','f','fi','o','oi','s','si']]
other_rep_s_words = [[{'l0'},{'<r0>,<l1>'},{'r1'}], [{'l1'},{'<r1>,<l0>'},{'r0'}], [{'<l0>, <l1>'},{'<r0>, <r1>'}]]
previous = None
result=[]
superpose_result=[]
choice=IntVar()
choice.set(0)
len_n=IntVar()
len_n.set(0)
style = tkFont.Font(family='Courier', size=14)
transitivity_array = compute_transitivty_table()
shorter_representations = possible_s_words()
draw_grid(view, None, None)
view.bind("<Button-1>", click_on_box)

'''
Adding the frame to show s-words and probability information
'''
fr = Frame(root, height=200, width=2000)
fr.pack_propagate(0) # don't shrink
fr.place(x=5, y=610)

'''
Adding the s-words information
'''
Label(fr, text="* Click on the grid elements to see probablity information", font=("Courier", 12)).place(x=900, y=0)
s_words_info_label_text = StringVar()
s_words_info_label = Label(fr, textvariable=s_words_info_label_text, font=("Courier", 13))
s_words_info_label.place(x=0, y=20)

'''
Adding the probablity information
'''
prob_info_label_text = StringVar()
prob_info_label = Label(fr, textvariable=prob_info_label_text, font=("Courier", 13))
prob_info_label.place(x=0, y=20)

root.mainloop()
```
